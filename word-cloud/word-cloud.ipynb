```python
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
from wordcloud import WordCloud
%matplotlib inline
```

```python
random_text = "This is random text with some repeating words like is random This repeating itsef is and is".split(" ")
```

```python
dictionary = FreqDist(random_text)
del dictionary[' ']
dictionary
```

```python
wcloud = WordCloud().generate_from_frequencies(dictionary)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wcloud, interpolation='bilinear')
plt.show()
```
